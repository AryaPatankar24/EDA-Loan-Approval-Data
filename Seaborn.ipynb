```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("/content/loan_approval_data (1).csv")
```

```python
df.head()
```

#Q1. How do you create a count plot for Loan Approval?

```python
sns.countplot(x= "Loan_Approved", data= df, palette= "magma")
plt.title("Loan Approval Count")
plt.show()
```

#Q2. How do you visualize Credit Score vs Loan Approval?

```python
df.groupby("Loan_Approved")["Credit_Score"].mean()*100
```

```python
sns.boxplot(x="Loan_Approved", y= "Credit_Score", data= df, palette= "PiYG")
plt.title("Credit Score vs Loan Approval")
plt.show()
```

#Q3. How do you compare Loan Amount across Property Areas?

```python
sns.boxplot(data= df,x="Property_Area", y="Loan_Amount", palette= "OrRd", hue= "Gender")
plt.title("Loan Amount by Property Area")
plt.show()
```

#Q4. How do you create a scatter plot with approval status

```python
sns.scatterplot(
    x= "Applicant_Income",
    y= "Loan_Amount",
    hue= "Loan_Approved",
    data =df
)
plt.title("Income vs Loan Amount by Approval Status")
plt.show()
```

#Q5. How do you visualize correlation using a heatmap?

```python
features= ["Applicant_Income", "Credit_Score", "Total_Income", "Saving_to_Loan_Ratio", "Loan_to_Income_Ratio"]
```

```python
numeric_df = df[features].select_dtypes(include=np.number)
corr_matrix = numeric_df.corr()
```

```python
sns.heatmap(corr_matrix, annot= True)
plt.title("Correlation Heatmap")
plt.show()
```

#Q6. How to create a boxplot of Loan Amount by Employment Status?

```python
sns.boxplot(data= df, x= "Employment_Status", y= "Loan_Amount", palette= "Purples")
plt.title("Loan Amount by Employment Status")
plt.show()
```

#Q7. How to create a violin plot of Credit Score?

```python
sns.violinplot(data= df, x= "Credit_Score", y= "Loan_Approved")
plt.title("Credit Score vs Loan Approval")
plt.show()
```

#Q8. How to plot DTI Ratio vs Loan Amount with size & color?

```python
sns.scatterplot(data= df, x= "DTI_Ratio", y= "Loan_Amount", size= "Savings", hue= "Existing_Loans", palette= "pastel")
plt.title("DTI Ratio vs Loan Amount")
plt.show()
```
