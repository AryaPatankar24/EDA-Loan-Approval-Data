```python
# Replace missing values
#Numeric columns
```

```python
df["Applicant_Income"].fillna(df["Applicant_Income"].median(), inplace=True)
df["Credit_Score"].fillna(df["Credit_Score"].median(), inplace= True)
```

```python
#Categorical columns
```

```python
df["Employment_Status"].fillna(df["Employment_Status"].mode()[0], inplace=True)
```

```python
#Creating New Professional Columns
```

```python
df["Total_Income"]= df["Applicant_Income"] + df["Coapplicant_Income"]
```

```python
df["Approval_Flag"]= df["Loan_Approved"].map({"Y": 1, "N": 0})
```

```python
#Credit Score Category(Using Conditions)
```

```python
def credit_category(score):
  if score >= 750:
    return "Excellent"
  elif score >= 650:
    return "Good"
  elif score >= 550:
   return "Average"
  else:
    return "Poor"
```

```python
df["Credit_Score_Category"] = df["Credit_Score"].apply(credit_category)
```
