#Q1. How would you fill missing numerical values professionally?

```python
num_cols= df.select_dtypes(include="number").columns
```

```python
df[num_cols]= df[num_cols].fillna(df[num_cols].median())
```

#Q2. How do you fill missing categorical values without bias?

```python
cat_cols = df.select_dtypes(include="object").columns
df[cat_cols] = df[cat_cols].fillna("Unknown")
```

#Q3. How do you create Total_Income using Pandas?

```python
df["Total_Income"]= df["Applicant_Income"]+ df["Coapplicant_Income"]
```

#Q4.How do you create a binary Approval_Flag column?

```python
df["Approval_Flag"] = df["Loan_Approved"].map({"Yes":1, "No": 0})
```

#Q5. How do you categorize Credit_Score in Pandas?

```python
df["Credit_Score_Category"]= pd.cut(
    df["Credit_Score"],
    bins=[0, 600, 700, 850],
    labels= ["Poor", "Average", "Good"]
)
```

#Q6. How do you derive Employment_Stability from Employment_Status?

```python
df["Employment_Stability"] = df["Employment_Status"].apply(
    lambda x: "Stable" if x == "Salaried" else "Variable"
)
```

#Q7. How do you analyze loan approval rate by Employment Stability?

```python
df.groupby("Employment_Stability")["Approval_Flag"].mean()
```

#Q8. How do you analyze approval trends by Property_Area?

```python
df.groupby("Property_Area")["Approval_Flag"].mean().sort_values(ascending= False)
```

#Q9. How do you filter high-risk loan applicants using Pandas?

```python
high_risk= df[
    (df["Credit_Score"] < 600) &
    (df["DTI_Ratio"] > 40)
]
```

#Q10. How do you perform correlation analysis in Pandas?

```python
df.corr(numeric_only=True)
```

#Q11. How do you summarize loan approvals by Credit_Score_Category and Loan_Purpose together?

```python
pd.pivot_table(
    df,
    values="Approval_Flag",
    index= "Credit_Score_Category",
    columns= "Loan_Purpose",
    aggfunc= "mean"
)
```

#Q12. How do you sort applicants by Loan_Amount?

```python
df.sort_values(by= "Loan_Amount", ascending= False)
```

#Q13. How do you export the cleaned dataset back to Excel?

```python
df.to_excel("cleaned_loan_data.xlsx", index=False)
```
